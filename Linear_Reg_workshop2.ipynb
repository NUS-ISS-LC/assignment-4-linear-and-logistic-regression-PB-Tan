# Linear Regression Example 

In this example, we will build a linear regression with two independent variables.  This linear regression model is also called multiple linear regression.
1) Load the dataset 'OmniPower.csv'
2) Create a Linear Regression Model with two independent variables (Price and Promotion), one dependent variable (Sale)
3) Provide input values and predict the Sale
4) What is the R square value?

```python
# 1) Load the dataset 'OmniPower.csv'
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('OmniPower.csv')
df.head()

```

```python
#visualize the relationship to develop an intuitive understanding

plt.scatter(df.Price, df.Sales, color = 'blue', label = 'Price')
# plt.scatter(df.Promotion, df.Sales, color = 'red')
plt.xlabel('Price or Promotion')
plt.ylabel('Sales')
```

```python
# 2) Create a Linear Regression Model with two independent variables (Price and Promotion), one dependent variable (Sale)
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(df[['Price','Promotion']], df.Sales, test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression

linReg = LinearRegression()
linReg.fit(x_train, y_train)

m = linReg.coef_
c = linReg.intercept_

print(f"Multivariate regression Eqn (in $): \nSales = {m[0]:.3f}*Price + {m[1]:.3f}*Promo + {c:.3f}")
```

```python
#3) Provide input values and predict the Sale
sample = [[60,400]]
y_pred = linReg.predict(sample)
print(f"Predicted Sale for price of ${sample[0][0]} and promo of ${sample[0][1]} is ${y_pred[0]:.3f}")
```

```python
# 4) What is the R square value?
from sklearn.metrics import r2_score

# overwrite y-hat with predicted y values in x-test
y_pred = linReg.predict(x_test)
score = r2_score(y_test, y_pred)
print(f'r2_score: {score:.5f}') 
```
